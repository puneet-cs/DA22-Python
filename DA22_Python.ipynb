# Define Variables in Python

```python
# Interview Question: Do python have data types or not?
# YES data types are present in python but we don't need to mention them.

a = 10  # int
b = 2.5 # float
c = "Hello Babai, How is Pinni"
d = True
e = None  # later on you can store any value in e

print(type(a),a)
print(type(b),b)
print(type(c), c)
print(type(d), d)
print(type(e), e)
```
