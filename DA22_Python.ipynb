# Match case statement

```python
# Match case statement is used to replace multiple if-else conditions.
# Match case is similar to Switch case in c & c++.
# Match case is more advance than the Switch case.
# Match case is having advance features like handling of data structure and conditions.
```

```python
# Example-1: structue of match case

num = int(input("Enter a number"))

match num:
    case 10:
        print("Hello")
    case 20:
        print("World")
    case _:
        print("something else")
```

```python
# Example-2 : match case with data structure
data = [10,20,30]

match data:
    case [1,2,3]:
        print("Hello-1")
    case [30,20,10]:
        print("Hello-2")
    case [10,20,30]:
        print("Hello-3")
    case _:
        print("Nothing Match")
```

```python
# Example-3: Match case using condition

num = int(input("Enter a number"))

match num:
    case x if x > 0:
        print("It is positive number")
    case y if y < 0:
        print("It is negative number")
    case _:
        print("It is ZERO")
```

```python
# Interview Question
x = 10
if x > 3:
    print("A")
    if x%3 == 0 and x:
        print("B")
    elif x == 15:
        print("D")
    else:
        pass
else:
    print("E")
```

# while loop

```python
# Example-1: structure of while loop
i = 1
while i <= 10:
    i += 1
    print(i)
```

```python
# Example-2: break statement while loop
# break statement is used to come out of the loop
i = 1
while i <= 10:
    print(i)
    if i == 5:
        break
    i += 1
```

```python
# Example-3: continue statement
# continue statement take the control to the start of the loop
# after executing continue statement, further next statement will skip

i = 1
while i <= 10:
    i += 1
    if i == 5 or i == 8:
        continue
    print(i)
```
