# Type casting

```python
# Convert one datatype to another datatype

# interger type casting : int()
v1 = int(2.8) # float to int
v2 = int("100")  # str to int
v3 = int(False)  # boolean to int
v4 = int("101", 2)  # Binary to int
v5 = int("C", 16) # hexadecimal to int
print(v1, type(v1))
print(v2, type(v2))
print(v3, type(v3))
print(v4, type(v4))
print(v5, type(v5))
```

```python
# float type casting : float()

v6 = float(10) # int to float
v7 = float("10.8") # str to float
v8 = float(False)  # boolean to float

print(v6, type(v6))
print(v7, type(v7))
print(v8, type(v8))
```

```python
# string type casting : str()
v9 = str(100) # int to string
v10 = str(3.5) # float to string
v11 = str(True)  # bool to str

print(v9, type(v9))
print(v10, type(v10))
print(v11, type(v11))
```
